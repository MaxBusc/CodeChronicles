# Mission 3

```
wildcodeschool =	{
  "data_analyst": 4,
  "data_scientist": 2,
  "dev_web": "13"
}


wildcodeschool ["dev_web"] = int (wildcodeschool["dev_web"])
wildcodeschool.update({"total_campus": 19})
print(wildcodeschool)
```

# Mission 5

# Mission 6

Create a script that returns the same elements as the **values()** method, as a list.

Example: for `{'a': 1, 'b': 2, 'c': 3}` it will create `[1, 2, 3]`

```
#using a for loop here is not necessary
my_dict = {"Test": 21, "Z": "alphabet", "Thre": 3, "D": 4,}

my_list = []

for i in my_dict.values():
    my_list.append(i)

print(my_list)
print(type(my_list))
```

```
#This answer is simpler and more effective !
my_dict = {"Test": 21, "Z": "alphabet", "Thre": 3, "D": 4,}
my_list=list(my_dict.values())
print(my_list)
print(type(my_list))
```
