Remember, using a for loop on a string will browse all the individual characters of it!

# Mission 6

Ask the user for a string. Then, display the same string with all vowels converted to uppercase, and all consonants converted to lowercase.  
For example:
* string = `"Papa's got a brand new bag"`  
Result : `"pApA's gOt A brAnd nEw bAg"`


````
vowels=["a","e","i","o","u","y"]
consonants=["c","d","f","g","h","j","k","l","m","n","p","q","r","s","t","v","w","x","z"]

string = input("Enter a string: ")
new_string = ""  # Pour stocker la nouvelle chaîne modifiée que l'on utilisera par la suite dans nos boucles

# Parcourir chaque caractère dans la chaîne
for char in string:
    if char.lower() in vowels:  # Si c'est une voyelle, la mettre en majuscule
        new_string += char.upper()
    elif char.lower() in consonants:  # Si c'est une consonne, la mettre en minuscule
        new_string += char.lower()
    else:
        new_string += char  # Si ce n'est ni voyelle ni consonne (espace, symbole...), on ne change pas

print("Result : ", new_string)
````
